# 01 — EDA: Coffee Futures Prices

**Objective**: Explore the distribution, trends, and volatility of ICE Arabica Coffee futures (KC=F).

**Data**: Daily OHLCV from Yahoo Finance via `yfinance`, cached locally as parquet.

**Date range**: 2021-01-01 to present.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd()))

import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

from src.data.fetch_coffee import fetch_coffee
from src.visualization.price_charts import (
    candlestick_chart,
    returns_histogram,
    rolling_volatility_chart,
    drawdown_chart,
    sma_chart,
)

df = fetch_coffee(start="2021-01-01")
print(f"Loaded {len(df)} trading days ({df.index.min()} to {df.index.max()})")
print(f"Columns: {list(df.columns)}")
df.head()
```

## 1.1 Summary Statistics

```python
close = df["Close"]
returns = close.pct_change().dropna() * 100

stats = pd.DataFrame({
    "Metric": ["Latest Price", "Mean", "Median", "Std Dev", "Min", "Max", "Skew", "Kurtosis"],
    "Close (¢/lb)": [
        f"{close.iloc[-1]:.2f}",
        f"{close.mean():.2f}",
        f"{close.median():.2f}",
        f"{close.std():.2f}",
        f"{close.min():.2f}",
        f"{close.max():.2f}",
        f"{close.skew():.2f}",
        f"{close.kurtosis():.2f}",
    ],
    "Daily Return (%)": [
        "", f"{returns.mean():.3f}", f"{returns.median():.3f}",
        f"{returns.std():.3f}", f"{returns.min():.2f}", f"{returns.max():.2f}",
        f"{returns.skew():.2f}", f"{returns.kurtosis():.2f}",
    ],
})
stats
```

## 1.7 Autocorrelation of Daily Returns

Check for autocorrelation in daily returns — if significant, the market has short-term predictability.

```python
from scipy import stats
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 4))
plot_acf(returns.dropna(), lags=30, ax=ax)
ax.set_title("Autocorrelation of Daily Returns")
plt.tight_layout()
plt.show()
```
